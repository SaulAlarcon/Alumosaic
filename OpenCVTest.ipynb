```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
import random
%matplotlib inline
```

```python
image = cv2.imread('can_images/aha_raw.jpg',cv2.IMREAD_UNCHANGED)
```

```python
def show_image(image):
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.show()
```

```python
show_image(image)
```

```python
image = cv2.rotate(image, cv2.cv2.ROTATE_90_COUNTERCLOCKWISE)
```

```python
show_image(image)
```

```python
# Convert to Grayscale for Contour Detection

image_grey = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
show_image(image_grey)
```

```python
# Apply Binary Thresholding

ret, thresh = cv2.threshold(image_grey, 150, 255, cv2.THRESH_BINARY)
```

```python
show_image(thresh)
```

```python
contours, hierarchy = cv2.findContours(image = thresh, mode = cv2.RETR_TREE, method = cv2.CHAIN_APPROX_NONE)

# Also need to take a look at RETR_CCOMP for 
```

```python
image_copy = image.copy()
```

```python
cv2.drawContours(image = image_copy, contours = contours, contourIdx=-1, color = (0,255,0), thickness = 3, lineType=cv2.LINE_AA)

show_image(image_copy)
```
